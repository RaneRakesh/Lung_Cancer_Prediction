### 12. Skewness

Skewness is a measure of the asymmetry of a distribution.

A distribution is asymmetrical when its left and right side are not mirror images. 

A distribution can have right (or positive), left (or negative), or zero skewness.

```python
def right_nor_left(df, int64_lst):
    
    temp_skewness = ['column', 'skewness_value', 'skewness (+ve or -ve)']
    temp_skewness_values  = []
    
    temp_total = ["positive (+ve) skewed", "normal distrbution" , "negative (-ve) skewed"]
    positive = 0
    negative = 0
    normal = 0
    
    for value in int64_lst:
    
        rs = round(df[value].skew(),4)

        if rs > 0:
            temp_skewness_values.append([value,rs , "positive (+ve) skewed"])   
            positive = positive + 1

        elif rs == 0:
            temp_skewness_values.append([value,rs,"normal distrbution"])
            normal = normal + 1

        elif rs < 0:
            temp_skewness_values.append([value,rs, "negative (-ve) skewed"])
            negative = negative + 1
    
    skewness_df =  pd.DataFrame(temp_skewness_values, columns=temp_skewness) 
    skewness_total_df =  pd.DataFrame([[positive, normal, negative]], columns=temp_total) 
    
    return skewness_df, skewness_total_df
```

```python
int64_cols = ['int64']
int64_lst_col = list(df.select_dtypes(include=int64_cols).columns)

skew_df,skew_total_df = right_nor_left(df, int64_lst_col)
```

```python
skew_df
```

```python
skew_total_df
```

**We notice with the above results that we have following details:**
- 2 columns are positive (+ve) skewed.
- 0 columns are normal distrbution.
- 13 columns are negative (-ve) skewed.

### 1. Now we spearate the target variable and feature columns in two different dataframe and will check the shape of the dataset for validation purpose.


```python
# Spearate target and feature column in X and y variable
target = 'LUNG_CANCER'
X = df.drop(target,axis=1)
y = df[target]
```

```python
# Check the shape of X and y variable
X.shape, y.shape
```

```python
# Reshape the y variable 
y = y.values.reshape(-1,1)
```

```python
# Again check the shape of X and y variable
X.shape, y.shape
```

### 2. Target Distribution Imbalance Handling

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN(random_state=42)
X, y = adasyn.fit_resample(X, y)
```

### 3. Split dataset into train and test dataset


Here we are spliting our dataset into 80/20 percentage where 80% dataset goes into the training and 20% goes into testing the part.

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y,
test_size= 0.25,
random_state=0)
```

```python
#Checking the shape of training and testing data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Reshaping of y_train
y_train = y_train.reshape(-1,1)
```

```python
y_train.shape
```

```python
#Reshaping of y_test
y_test = y_test.reshape(-1,1)
```

```python
y_test.shape
```

```python
#Rechecking the shape of training and testing data
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 4. In this step we are going to train our dataset on different Machine Learning algorithms. As we know that our target variable is not in continous format so we have to apply classification algorithms.

**Algorithms we are going to use in this step are:**

1) Logistic Regression

2) Decision Tree

3) Random Forest

 **1) Logistic Regression**

```python
#Fitting training data to the model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(random_state=0)
model.fit(X_train, y_train)
```

```python
#Predicting result using testing data
y_pred= model.predict(X_test)
y_pred
```

```python
#Model accuracy
from sklearn.metrics import classification_report, accuracy_score, f1_score
print(classification_report(y_test, y_pred))
```

### This model is almost 98% accurate.

**2) Decision Tree**

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
#Fitting training data to the model
dtree = tree.DecisionTreeClassifier()
dtree = dtree.fit(X_train,y_train)
```

```python
#Predicting result using testing data
y_predicted = dtree.predict(X_test)
```

```python
#Model accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_predicted)*100
print("Accuracy of Decision Tree is: ",acc)
```

### This model is almost 95.83% accurate.

**3) Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran_clf = RandomForestClassifier(max_depth=2, random_state=0)
```
